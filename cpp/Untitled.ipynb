```c++
double Normal(double x) {
    return (exp(-(pow(x,2))/2)/(sqrt(2*M_PI)));
}
```

```c++
double estadoAtual, epsilon, custoAtual, candidato, custoCandidato, sorteio;
int n, idx, saida;
std::random_device rd;
std::mt19937 gen(rd());
std::uniform_real_distribution<double> uniform(0,1);

saida = 2000;
estadoAtual = 1;
epsilon = 1;

custoAtual = Normal(estadoAtual);

for (n = 0; n < N; n++){
    candidato = estadoAtual + epsilon*(2*uniform(gen)-1);
    custoCandidato = Normal(candidato);
    sorteio = uniform(gen);
    if (sorteio < custoCandidato/custoAtual){
        estadoAtual = candidato;
        custoAtual = custoCandidato;
    }
    if (N-n < saida){
        cout << custoAtual << ',';
    }
}
```
